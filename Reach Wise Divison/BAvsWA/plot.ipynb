```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import openpyxl
```

```python
input_folder = 'F:\One Drive\OneDrive - Indian Institute of Science Education and Research Bhopal\Excel data for Various Parameters\Reach Wise Divison\BAvsWA'
df = pd.read_excel('Density.xlsx',sheet_name='Sheet1')
df2 = pd.read_excel('Density.xlsx',sheet_name='Sheet2')
df3 = pd.read_excel('Density.xlsx',sheet_name='Sheet3')
df4 = pd.read_excel('Density.xlsx',sheet_name='Sheet4')
```

```python
plt.plot(df['Reach'], df['Density-Pre-2017'], color='red',
         marker='o', linewidth=2, markersize=5,label='Pre-2017')
plt.plot(df2['Reach'], df2['Density-Mon-2017'],
         color='blue', marker='o', linewidth=2, markersize=5,label='Mon-2017')
plt.plot(df3['Reach'], df3['Density-Post-2017'],
         color='green', marker='o', linewidth=2, markersize=5,label='Post-2017')
plt.plot(df4['Reach'], df4['Density-1965'],
            color='black', marker='o', linewidth=2, markersize=5,label='1965')








plt.xlabel('Reach')
plt.ylabel('Density as (BA/CA)')
plt.legend()
title = 'Bar Density 1965'
plt.title(title, fontdict={'fontname': 'Times New Roman'})
plt.grid(True, color='grey', linestyle='-', linewidth=0.25, alpha=0.5)

plt.show()
```
