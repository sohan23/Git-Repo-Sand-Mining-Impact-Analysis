```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import openpyxl
```

```python
# Read the excel file
df = pd.read_excel('F:\One Drive\OneDrive - Indian Institute of Science Education and Research Bhopal\Excel data for Various Parameters\Reach Wise Divison\BAvsWA\concatenated_file-Plotting.xlsx')
df
```

```python
# Plot the 1965 Exposed Bar vs Water Area
x = df.iloc[:, 0]
A = df['Pre-2017-Water']
B = df['Mon-2017-Water']
C = df['Post-2017-Water']

bar_width = 0.18
x1 = np.arange(len(x))
x2 = [i + bar_width for i in x1]
x3 = [i + bar_width for i in x2]

plt.figure(figsize=(16, 10))  # increase figure size
plt.bar(x1, A, width=bar_width, label='Pre-2017-Water')
plt.bar(x2, B, width=bar_width, label='Mon-2017-Water')
plt.bar(x3, C, width=bar_width, label='Post-2017-Water')

# Add vertical text annotations with values rounded to 3 decimal points and increased spacing
for i in range(len(A)):
    plt.text(x=x1[i]-0.05, y=A[i]+0.01, s=f"{A[i]:.3f}", size=9, rotation=90)
    plt.text(x=x2[i]-0.05, y=B[i]+0.01, s=f"{B[i]:.3f}", size=9, rotation=90)
    plt.text(x=x3[i]-0.05, y=C[i]+0.01, s=f"{C[i]:.3f}", size=9, rotation=90)

plt.xticks([i + bar_width for i in range(len(x))], x)
plt.grid(True, color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
plt.xlabel('Reach')
plt.ylabel('Channel Area')
plt.title('2017 Variability in Channel Area', fontweight='bold',
          fontdict={'fontname': 'Times New Roman'}, fontsize=18)
plt.legend()
plt.show()
```
