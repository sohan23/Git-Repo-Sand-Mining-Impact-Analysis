# RivWidth Calculation with python

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import os
import openpyxl
from scipy.interpolate import splrep, splev
%matplotlib inline
```

## Defining the Input folder

```python
input_folder = 'F:\One Drive\OneDrive - Indian Institute of Science Education and Research Bhopal\Excel data for Various Parameters\River Width'
```

```python
#Read the excel data
df = pd.read_excel(os.path.join(input_folder, 'RivWidth-1965.xls'))
df = df.iloc[2:]
df
```

```python

# Create a new column 'marker' to control the marker style of each point in the scatter plot
df['marker'] = df.groupby('Range')['Length'].transform(
    lambda x: ['o' if i == 1 else '^' for i in x.rank(ascending=False)])

# Create a scatter plot with x as Range and y as Length, where the marker value controls the marker style of each point
for marker in df['marker'].unique():
    plt.scatter(df.loc[df['marker'] == marker, 'Range'],
                df.loc[df['marker'] == marker, 'Length'], marker=marker,s=36)

# Show the plot
plt.xlabel('Range')
plt.ylabel('Length (m)')
plt.title('River Width-1965')
plt.legend(['Primary Channel', 'Secondary Channel'], loc='best')
plt.grid(True, color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
plt.show()
```
