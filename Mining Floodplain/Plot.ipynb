```python
import numpy as np
import pandas as pd
import os
# import matplotlib as mpl
import matplotlib.pyplot as plt
# import matplotlib.ticker as tck
%matplotlib inline
```

```python
input_folder ='F:\One Drive\OneDrive - Indian Institute of Science Education and Research Bhopal\Excel data for Various Parameters\Mining Floodplain'
```

```python
df = pd.read_excel(os.path.join(input_folder,'Mining-All Final.xlsx'))
df
```

```python
plt.plot(df['Area'],color='black',marker='o',linestyle='dashed',linewidth=2,markersize=5)
plt.xlabel('From Monsoon 2016 to Post Monsoon 2020')
plt.ylabel('Mining Area in sq. km')
plt.title('Mining Area variability in Floodplain')
plt.figure(figsize=(16,10))
plt.show()
```

```python
df.iloc[0:5].plot(x='Year', y='Area',
                  color='black', marker='o', linewidth=2, markersize=5)
plt.xlabel('Season of the Year')
plt.ylabel('Mining Area in sq. km')
plt.title('Mining Area variability in Floodplain')
df.iloc[-9:10].plot(x='Year', y='Area',
                  color='red', marker='s', linewidth=2, markersize=5)
plt.xlabel('Season of the Year')
plt.ylabel('Mining Area in sq. km')
plt.title('Mining Area variability in Floodplain')
df.iloc[-4:15].plot(x='Year', y='Area',
            color='magenta', marker='o',  linewidth=2, markersize=5)            
plt.xlabel('Season of the Year')
plt.ylabel('Mining Area in sq. km')
plt.title('Mining Area variability in Floodplain')
plt.figure(figsize=(16, 10))
plt.show()
```

```python
df.iloc[0:5].plot.bar(x='Year', y='Area')
plt.xlabel('Season of the Year')
plt.ylabel('Mining Area in sq. km')
plt.title('Mining Area variability in Floodplain')
df.iloc[-9:10].plot.bar(x='Year', y='Area')
plt.xlabel('Season of the Year')
plt.ylabel('Mining Area in sq. km')
plt.title('Mining Area variability in Floodplain')
df.iloc[-4:15].plot.bar(x='Year', y='Area')
plt.xlabel('Season of the Year')
plt.ylabel('Mining Area in sq. km')
plt.title('Mining Area variability in Floodplain')
plt.figure(figsize=(16, 10))
plt.show()
```
